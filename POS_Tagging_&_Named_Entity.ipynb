```python
corpus = ''' I have a dream that one day this nation will rise up and live out the true meaning of its creed: "We hold these truths to be self-evident, that all men are created equal."

I have a dream that one day on the red hills of Georgia, the sons of former slaves and the sons of former slave owners will be able to sit down together at the table of brotherhood.

I have a dream that one day even the state of Mississippi, a state sweltering with the heat of injustice, the state sweltering with the heat of oppression, will be transformed into an oasis of freedom and justice.

'''
```

### ***Tokenization***

---



---



```python
import nltk
nltk.download('punkt')
```

```python
### SENTENCE TOKENIZER

from nltk.tokenize import sent_tokenize

document = sent_tokenize(corpus)
```

```python
document
```

```python
### WORD TOKENIZER

from nltk.tokenize import word_tokenize

word_tokenizer = word_tokenize(corpus)
```

```python
word_tokenizer
```

### ***Stop Words & Post Tagging***

---



---



```python
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')


for word in word_tokenizer:
  words = [word for word in word_tokenizer if word not in set(stopwords.words('english'))]
  tag = nltk.pos_tag(words)
print(tag)
```
